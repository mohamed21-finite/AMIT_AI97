```python
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speaks(self):

        print("person {self.name} speaks")

    
class doctor(person):
    def speaks(self):
        print(f"doctor {self.name} speaks")

doctor1 = doctor("ahmed", 21)

doctor1.speaks()



```

```python
class student:
    def __init__(self, name, grade):
        self.name = name 
        self.grade = grade

    def update_grade(self, new_grade):
        self.grade = new_grade

    def show_info(self):
        print(f"student name {self.name}, grade: {self.grade}")

student1 = student("mohamed", 90)
student1.show_info()

student1.update_grade(95)

student1.show_info()
```

```python
class bird:
    def __init__(self, wing_size):
        self.wing_size = wing_size

    def fly(self):
        print(f"i can fly, my wing size: {self.wing_size}")

class swimmer:
    def __init__(self, fin_size):
        self.fin_size = fin_size

    def swim(self):
        print(f"i can swim, my fin size: {self.fin_size}")

class Duck(bird, swimmer):
    def __init__(self, wing_size, fin_size):
        bird.__init__(self, wing_size)
        swimmer.__init__(self, fin_size)

    def get_info(self):
        print(f"I can swim, my fin size: {self.fin_size} and I can fly, my wing size: {self.wing_size}")


        

fish1 = swimmer(2)
fish1.swim()

bird1 = bird(2)
bird1.fly()

duck = Duck(1,1)
duck.swim()
duck.fly()
duck.get_info()
```

```python
class Child:
    def __init__(self, age, school):
        self.age = age
        self.school = school

    def info(self):
        print("I am a child")

class Student(Child):
    def __init__(self, age, school, grade):
        super().__init__(age, school)
        self.grade = grade

    def info2(self):
        super().info()
        print("I am a student")

student = Student(21, "college", 90)

student.info2()
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Mammal(Animal):
    def __init__(self, name):
        super().__init__(name)

    def eat(self):
        print(f"{self.name} eats")

class Dog(Mammal):
    def __init__(self, name):
        super().__init__(name)

    def walk(self):
        print(f"{self.name} walks")

dog = Dog("roy")

dog.eat()
dog.walk()
```

```python
class Mobile:
    def __init__(self, brand):
        self.brand = brand

    def power_on(self):
        print("mobile is running")
    
    def info(self):
        print(f"brand is {self.brand}")

class Phone(Mobile):
    def __init__(self,brand, number):
        Mobile.__init__(self, brand)
        self.number = number

    def make_call(self):
        print("calling")

class Camera(Mobile):
    def __init__(self,brand, pixel):
        Mobile.__init__(self, brand)
        self.pixel = pixel

    def take_photo(self):
        print("photo is taken")


class Smartphone(Camera, Phone):
    def __init__(self, brand, number, pixel):
        Phone.__init__(self, brand, number)
        Camera.__init__(self, brand, pixel)

    def info(self):
        print(f"brand: {self.brand}, number: {self.number}, pixel: {self.pixel}")

mobile = Smartphone("SAMSUNG", 1234, 8)

mobile.info()
mobile.take_photo()
mobile.make_call()
```
