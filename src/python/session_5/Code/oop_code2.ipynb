## OOP

```python
# blue print
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_info(self):
        print(f"name: {self.name}, age: {self.age}")


person1 = person("ahmed", 21)

person1.get_info()
```

```python
class dog:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print("bark")

dog1 = dog("roy")
dog2 = dog("troy")

dog1.sound()
```

```python
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speaks(self):

        print("person {self.name} speaks")

    
class doctor(person):
    def speaks(self):
        print(f"doctor {self.name} speaks")

doctor1 = doctor("ahmed", 21)

doctor1.speaks()



```

```python
class student:
    def __init__(self, name, grade):
        self.name = name 
        self.grade = grade

    def update_grade(self, new_grade):
        self.grade = new_grade

    def show_info(self):
        print(f"student name {self.name}, grade: {self.grade}")

student1 = student("mohamed", 90)
student1.show_info()

student1.update_grade(95)

student1.show_info()
```

```python
class bird:
    def __init__(self, wing_size):
        self.wing_size = wing_size

    def fly(self):
        print(f"i can fly, my wing size: {self.wing_size}")

class swimmer:
    def __init__(self, fin_size):
        self.fin_size = fin_size

    def swim(self):
        print(f"i can swim, my fin size: {self.fin_size}")

class Duck(bird, swimmer):
    def __init__(self, wing_size, fin_size):
        bird.__init__(self, wing_size)
        swimmer.__init__(self, fin_size)

    def get_info(self):
        print(f"I can swim, my fin size: {self.fin_size} and I can fly, my wing size: {self.wing_size}")


        

fish1 = swimmer(2)
fish1.swim()

bird1 = bird(2)
bird1.fly()

duck = Duck(1,1)
duck.swim()
duck.fly()
duck.get_info()
```
