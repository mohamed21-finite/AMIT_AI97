```python
# -------------------------------------------------------------
# IMPORT REQUIRED LIBRARIES
# -------------------------------------------------------------

# numpy → Used for numerical operations (arrays, math functions)
import numpy as np

# pandas → Used for working with datasets (tables of data)
import pandas as pd

# matplotlib.pyplot → Main plotting module in Matplotlib
import matplotlib.pyplot as plt

# seaborn → High-level plotting library built on top of Matplotlib
import seaborn as sns

# This magic command tells Jupyter Notebook to show plots
# directly inside the notebook (instead of a separate window)
%matplotlib inline

# Optional: Set the default style for Seaborn plots
# style="whitegrid" means background is white with gridlines
sns.set_theme(style="whitegrid")
```

```python
# -------------------------------------------------------------
# MULTILINE PLOT
# -------------------------------------------------------------

# 1. CREATE THE DATA
# ------------------------------

# Days of the week (X-axis labels)
days = ["Sat", "Sun", "Mon", "Tue", "Wed", "Thu", "Fri"]

# Example temperatures (°C) for Cairo
cairo_temps = [30, 32, 31, 29, 28, 27, 26]

# Example temperatures (°C) for Alexandria
alex_temps = [25, 26, 27, 28, 28, 29, 30]

# Example temperatures (°C) for Alexandria
sharm_temps = [33, 33, 34, 30, 29, 28, 27]

# 2. CREATE THE FIGURE
# ------------------------------

# Create a new figure (canvas) with specific size in inches
plt.figure(figsize=(8, 5))

# 3. PLOT THE FIRST LINE (Cairo)
# ------------------------------

# days → X-axis values
# cairo_temps → Y-axis values
# label → name for legend
# color → line color
# linewidth → line thickness
plt.plot(days, cairo_temps, label="Cairo", color="orange", linewidth=2)

# 4. PLOT THE SECOND LINE (Alexandria)
# ------------------------------

# linestyle="--" → dashed line for contrast
plt.plot(days, alex_temps, label="Alexandria", color="blue", linestyle="--")

# 5. PLOT THE SECOND LINE (Sharm)
# ------------------------------

# linestyle="--" → dashed line for contrast
plt.plot(days, sharm_temps, label="Sharm", color="red", linewidth=2)

# 6. ADD TITLES AND LABELS
# ------------------------------

plt.title("Weekly Temperature Comparison between Cities")  # Main chart title
plt.xlabel("Day of the Week")  # X-axis label
plt.ylabel("Temperature (°C)")  # Y-axis label

# 7. ADD LEGEND AND GRID
# ------------------------------

plt.legend()       # Show legend box
plt.grid(True)     # Show grid lines

# 8. DISPLAY THE PLOT
# ------------------------------

plt.show()  # Render the plot
```

```python
# -------------------------------------------------------------
# BAR CHART
# -------------------------------------------------------------

# Define category names
products = ["TV", "Mobile", "AC", "Fan"]

# Define values for each category
quantities = [5, 7, 3, 4]

# Create a new figure (6 inches wide × 4 inches tall)
plt.figure(figsize=(6, 4))

# Create a bar chart
# color= [...] → assign a different color to each bar
plt.bar(products, quantities,
        color=["skyblue", "salmon", "lightgreen", "orange"])

# Add title
plt.title("Bar Chart Example")

# Add x-axis label
plt.xlabel("Products")

# Add y-axis label
plt.ylabel("Quantities")

# Show the plot
plt.show()
```

```python
# -------------------------------------------------------------
# SCATTER PLOT 
# -------------------------------------------------------------

# For reproducibility (same random numbers every run)
np.random.seed(0)

# Generate 50 random numbers for x and y (between 0 and 1)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Create figure
plt.figure(figsize=(6, 4))

# Create scatter plot
# alpha=0.7 → makes points slightly transparent
plt.scatter(x_scatter, y_scatter, color="purple", alpha=0.7)

# Add title and labels
plt.title("Scatter Plot Example")
plt.xlabel("Random X")
plt.ylabel("Random Y")

# Show plot
plt.show()
```

```python
# -------------------------------------------------------------
# HISTOGRAM
# -------------------------------------------------------------

# Generate 1000 random numbers from a normal distribution
data = np.random.randn(1000)

# Create figure
plt.figure(figsize=(6, 4))

# Create histogram
# bins=20 → divide the range into 20 intervals
# edgecolor="black" → black edges around bars
plt.hist(data, bins=20, color="steelblue", edgecolor="black")

# Add title and labels
plt.title("Histogram Example")
plt.xlabel("Data")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

```python
# -------------------------------------------------------------
# PIE CHART
# -------------------------------------------------------------

# Percentages for each slice
sizes = [30, 20, 25, 25]

# Labels for each slice
fruits = ["Apples", "Bananas", "Cherries", "Dates"]

# Create figure
plt.figure(figsize=(6, 6))

# Create pie chart
# autopct="%1.1f%%" → show percentages with 1 decimal place
# colors → choose a pastel color palette from Seaborn
plt.pie(sizes, labels=fruits,
        autopct="%1.1f%%",
        colors=sns.color_palette("pastel"))

# Add title
plt.title("Pie Chart Example")

# Show plot
plt.show()
```

```python
# -------------------------------------------------------------
# LOAD EXAMPLE DATASET
# -------------------------------------------------------------

# Load Seaborn's built-in "tips" dataset
# It contains info about restaurant bills, tips, and customers
tips = sns.load_dataset("tips")

# Show first 5 rows of the dataset
tips.head()
```

```python
# -------------------------------------------------------------
# COUNTPLOT 
# -------------------------------------------------------------

# Count how many times each day appears in the dataset
# palette="viridis" → color scheme
plt.figure(figsize=(6, 4))
sns.countplot(
    x="day",
    data=tips,
    hue="smoker",          # 👈 match hue to x
    palette="viridis",  # color palette
    legend=False        # hide unnecessary legend
)
plt.title("Count of Records per Day")
plt.show()
```

```python
# -------------------------------------------------------------
# BOXPLOT
# -------------------------------------------------------------

# Show distribution of total_bill for each day
# Box shows median, quartiles, and outliers
plt.figure(figsize=(6, 4))
sns.boxplot(x="day", y="total_bill", data=tips, palette="coolwarm")
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

```python
# -------------------------------------------------------------
# VIOLIN PLOT
# -------------------------------------------------------------

# Similar to boxplot but shows full distribution (like a violin shape)
plt.figure(figsize=(6, 4))
sns.violinplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Violin Plot Example")
plt.show()
```

```python
# -------------------------------------------------------------
# HEATMAP
# -------------------------------------------------------------

# Keep only numeric columns for correlation
numeric_corr = tips.corr(numeric_only=True)

# Show correlation between numeric columns
plt.figure(figsize=(8, 5))
sns.heatmap(numeric_corr, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# -------------------------------------------------------------
# CUSTOMIZING STYLES
# -------------------------------------------------------------

# Change style to darkgrid
sns.set_style("darkgrid")

# Create histogram with KDE (smooth curve)
plt.figure(figsize=(6, 4))
sns.histplot(tips["total_bill"], kde=True, color="orange")
plt.title("Histogram with KDE - Darkgrid Style")
plt.show()
```

```python
data = {
    'Score': [85, 87, 90, 88, 300, 86, 89]
}

#[300]
#[85, 87, 90, 88, 86, 89]

df = pd.DataFrame(data)
print(df)
```

```python
# Detect outlier

Q1 = df['Score'].quantile(0.25)
Q3 = df['Score'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[ (df['Score'] < lower_bound) | (df['Score'] > upper_bound) ]
df_no_outliers = df[ (df['Score'] >= lower_bound) & (df['Score'] <= upper_bound) ]

print(outliers)
```
