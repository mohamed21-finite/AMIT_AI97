```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd


```

```python
# ===============================================
# CREATE DATA
# ===============================================

# Temperature data (for line, bar, and box plots)
days = ["Sat", "Sun", "Mon", "Tue", "Wed", "Thu", "Fri"]
cairo = [30, 32, 31, 29, 28, 27, 26]
alex = [25, 26, 27, 28, 28, 29, 30]
sharm = [33, 33, 34, 30, 29, 28, 27]

# Create a DataFrame for convenience (like a spreadsheet)
temp_df = pd.DataFrame({
    "Day": days,
    "Cairo": cairo,
    "Alexandria": alex,
    "Sharm El-Sheikh": sharm
})

# Simple sales dataset for scatter and bar charts
sales_df = pd.DataFrame({
    "Product": ["TV", "Mobile", "AC", "Fan"],
    "Units_Sold": [120, 80, 150, 60],
    "Revenue": [3000, 2200, 4500, 1800],
    "Region": ["North", "East", "South", "West"]
})
```

```python
# interactive line chart
```

```python
fig = go.Figure()
```

```python
fig.add_trace(go.Scatter(
    x=days,
    y=cairo,
    mode='lines+markers',
    name='Cairo',
    line=dict(color='orange', width=3)
))

fig.add_trace(go.Scatter(
    x=days,
    y=alex,
    mode='lines+markers',
    name='Alex',
    line=dict(color='blue', dash='dash', width=3)
))

fig.add_trace(go.Scatter(
    x=days,
    y=sharm,
    mode='lines+markers',
    name='Sharm',
    line=dict(color='red', width=3)
))

fig.update_layout(
    title="weekly Temperature Comparison",
    xaxis_title="Days of the week",
    yaxis_title="Temperature",
    hovermode="x unified",
    template="plotly_white",
    
)
```

```python
# ===============================================
# CREATE DATA
# ===============================================

# Temperature data (for line, bar, and box plots)
days = ["Sat", "Sun", "Mon", "Tue", "Wed", "Thu", "Fri"]
cairo = [30, 32, 31, 29, 28, 27, 26]
alex = [25, 26, 27, 28, 28, 29, 30]
sharm = [33, 33, 34, 30, 29, 28, 27]

# Create a DataFrame for convenience (like a spreadsheet)
temp_df = pd.DataFrame({
    "Day": days,
    "Cairo": cairo,
    "Alexandria": alex,
    "Sharm El-Sheikh": sharm
})

# Simple sales dataset for scatter and bar charts
sales_df = pd.DataFrame({
    "Product": ["TV", "Mobile", "AC", "Fan"],
    "Units_Sold": [120, 80, 150, 60],
    "Revenue": [3000, 2200, 4500, 1800],
    "Region": ["North", "East", "South", "West"]
})
```

```python
fig = px.bar(
    sales_df,
    x="Product",
    y="Revenue",
    color="Region",
    text="Revenue",
    title="Revenue by Product and Region",
    template="plotly_white"
)
```

```python
fig.update_traces(textposition='outside')
fig.update_layout(
    xaxis_title="Product Name",
    yaxis_title="Revenue",
    showlegend=True
)
```

```python
fig = go.Figure()
```

```python
fig.add_trace(go.Scatter(
    x = days ,
    y = cairo ,
    mode = 'lines+markers' ,
    name = 'cairo',
    line = dict(color = 'orange', width = 2)
))
#Line for alex temp

fig.add_trace(go.Scatter(
    x = days ,
    y = alex ,
    mode = 'lines+markers' ,
    name = 'alex',
    line = dict(color = 'red', width = 2)
))
#Line for sharm temp
fig.add_trace(go.Scatter(
    x = days ,
    y = sharm ,
    mode = 'lines+markers' ,
    name = 'sharm',
    line = dict(color = 'blue', width = 2)
))


fig.update_layout(
    updatemenus = [
        dict(
            type="dropdown",
            x=1.15, y=1.2,
            buttons=[
                dict(label="All Cities", method="update",
                     args=[{"visible": [True, True, True]}, {"title": "All cities"}]),
                dict(label="Cairo Only", method="update",
                     args=[{"visible": [True, False, False]}, {"title": "Cairo Only"}]),
                dict(label="Alex Only", method="update",
                     args=[{"visible": [False, True, False]}, {"title": "Alex Only"}]),  
                dict(label="Sharm Only", method="update",
                     args=[{"visible": [False, False, True]}, {"title": "Sharm Only"}])  
            ]
        )
    ],
    title="Interactive Temprature Selector",
    xaxis_title="Day",
    yaxis_title="Temprature",
    template="plotly_white"
)
```
