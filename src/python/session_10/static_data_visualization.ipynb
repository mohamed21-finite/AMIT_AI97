```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
# Muliline plot
```

```python
days = ["Sat", "Sun", "Mon", "Tue", "Wed", "Thu", "Fri"]
cairo_temps = [30, 32, 31, 29, 28, 27, 26]
alex_temps = [25, 26, 27, 28, 28, 29, 30]
sharm_temps = [33, 33, 34, 30, 29, 28, 27]
```

```python
# 1. Create figuer (Canvas)
```

```python
plt.figure(figsize=(8, 5))
```

```python
# Draw the plot lines
```

```python
plt.plot(days, cairo_temps, label="Cairo", color="orange", linewidth=2)
plt.plot(days, alex_temps, label="Alex", color="blue", linewidth=2, linestyle="--")
plt.plot(days, sharm_temps, label="Sharm", color="red", linewidth=2)

plt.title("Weekly Temerature Comparison between Cities")
plt.xlabel("Days of the Week")
plt.ylabel("Temerature")

plt.legend()
plt.grid(True)

plt.show()
```

```python
# Bar charts
```

```python
# Define category names
products = ["TV", "Mobile", "AC", "Fan"]

# Define values for each category
quantities = [5, 7, 3, 4]
```

```python
plt.figure(figsize=(6,4))
```

```python
plt.bar(products, quantities, color=["skyblue", "salmon", "lightgreen", "orange"])

plt.title("Products Quntities")
plt.xlabel("Products")
plt.ylabel("Quantities")

plt.show()
```

```python
# Scatter Plot
```

```python
np.random.seed(0) # in order to generate the same numbers every time
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
```

```python
plt.figure(figsize=(6, 4))
```

```python
plt.scatter(x_scatter, y_scatter, color="purple", alpha=0.7)

plt.title("Scatter Plot Example")
plt.xlabel("Random X")
plt.ylabel("Random Y")

plt.show()
```

```python
# Load tips dataset
tips = sns.load_dataset("tips")
tips
```

```python
plt.figure(figsize=(6,4))
```

```python
sns.countplot(
    x="day",
    data=tips,
    hue="smoker",
    palette="viridis",
    legend=True
)

plt.title("Count of Visits per Day")
plt.show()
```

```python
# Boxplot
```

```python
plt.figure(figsize=(6,4))
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="coolwarm")
plt.title("Boxplot")
plt.show()
```

```python
# Heatmap
```

```python
numeric_corr = tips.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,5))
```

```python
sns.heatmap(numeric_corr, annot=True, cmap="YlGnBu")

plt.title("Corrleation Heatmap")
plt.show()
```
