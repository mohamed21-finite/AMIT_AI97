## Pandas Series

What is a Series?

A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

````python
a1 = [1, 7, 2]

s1 = pd.Series(a1)

print(s1)
````

##### Labels:

If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.

This label can be used to access a specified value.

````python
print(s1[1])
````

## Create Labels:

With the index argument, you can name your own labels.

````python
a2 = [1, 7, 2]

s2 = pd.Series(a2, index = ["a", "b", "c"])

print(s2)
````

When you have created labels, you can access an item by referring to the label.

````python
print(s2["b"])
````

**Key/Value Objects as Series:**

You can also use a key/value object, like a dictionary, when creating a Series.

````python
calories = {"day1": 420, "day2": 380, "day3": 390}

s3 = pd.Series(calories)

print(s3)
````

Note: The keys of the dictionary become the labels.

To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series.

````python
a2 = [420, 380, 390]

s2 = pd.Series(a2, index = ["day1", "day2", "day3"])

print(s2)
````

````python
calories = {"day1": 420, "day2": 380, "day3": 390}

s4 = pd.Series(calories, index = ["day1", "day3"])

print(s4)
````

## DataFrames

What is a DataFrame?

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

````python
data1 = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

# Load data into a DataFrame object:
df1 = pd.DataFrame(data1)

print(df1) 
````

**Locate Row:**

As you can see from the result above, the DataFrame is like a table with rows and columns.

Pandas use the loc attribute to return one or more specified row(s)

````python
# Refer to the row index:

print(df1.loc[2])
````

Note: This example returns a Pandas Series.

````python
# Use a list of indexes:

print(df1.loc[[0, 2]])
````

Note: When using [], the result is a Pandas DataFrame.

**Named Indexes:**

With the index argument, you can name your own indexes.

````python
# Add a list of names to give each row a name:

data2 = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df2 = pd.DataFrame(data2, index = ["day1", "day2", "day3"])

print(df2) 
````

**Locate Named Indexes:**

Use the named index in the loc attribute to return the specified row(s).

````python
# Refer to the named index:

print(df2.loc["day2"])
````

**Filtering DataFrames:**

````python
# Create a sample DataFrame with custom index
data = {
    "Name": ["Ahmed", "Mohamed", "Hossam", "Heba", "Mostafa"],
    "Age": [25, 30, 35, 40, 28],
    "City": ["Cairo", "Alex", "Giza", "Luxor", "Aswan"]
}

df3 = pd.DataFrame(data, index=["a", "b", "c", "d", "e"])

print(df3)
````

- **.loc[] â€“ Label-based indexing:**

````python
# 1. Select single row by label:

print(df3.loc["b"])  # Row with index label "b"
````

````python
# 2. Select multiple rows by label list:

print(df3.loc[["a", "e"]])
````

````python
print(df3.loc["b"]) 
print(df3.loc[["a","b","d"]]) 
print(df3.loc["b":"d"]) 
````

````python
# 3. Select a range of rows (inclusive):

print(df3.loc["b":"d"])  # Includes "d"
````

````python
# 4. Select rows and columns by label:

print(df3.loc["c", "Age"])  # Cell at row "c", column "Name"
````

````python
# 5. Select multiple rows and columns:

print(df3.loc[["a", "d"], ["Name", "City"]])
````

````python
# 6. Conditional selection using .loc:

print(df3.loc[df3["Age"] > 30])  # All rows where Age > 30
````

````python
print(df3.loc[df3["Age"] == 30])  # Age equal to 30
````

````python
print(df3.loc[df3["Age"] != 30])  # Age not equal to 30
````

````python
print(df3.loc[(df3["Age"] > 25) & (df3["City"] == "Giza")])
````

````python
print(df3.loc[(df3["Age"] < 30) | (df3["City"] == "Luxor")])
````

````python
print(df3.loc[(df3["Age"] >= 28) & (df3["Age"] <= 35)])
````

````python
# Check if value exists in a list (using isin):

print(df3.loc[df3["City"].isin(["Cairo", "Giza"])])
````

- **.iloc[] â€“ Integer position-based indexing:**

````python
# 1. Select single row by position:

print(df3.iloc[1])  # Second row (Bob)
````

````python
# 2. Select a specific cell (row 2, column 1):

print(df3.iloc[2, 1])  
````

````python
# 3. Select a range of rows (like slicing):

print(df3.iloc[1:4])  # Rows 1 to 3 (Bob to David)
````

````python
# 4. Select rows and specific columns:

print(df3.iloc[[0, 3], [0, 2]])  # Rows 0 and 3, Columns 0 and 2
````

````python
# 5. Select entire row or column:

print(df3.iloc[:, 1])  
print(df3.iloc[2, :])  
````

````python
# 6. Negative indexing:

print(df3.iloc[-1])  # Last row (Mostafa)
````

**Merge DataFrame:**

````python
# Create first DataFrame: contains employee IDs and names
df_employees = pd.DataFrame({
    'EmployeeID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie']
})

# Create second DataFrame: contains employee IDs and departments
df_departments = pd.DataFrame({
    'EmployeeID': [101, 102, 104],
    'Department': ['HR', 'Finance', 'IT']
})
````

````python
# Merge the two DataFrames on the 'EmployeeID' column (inner join by default)
merged_df = pd.merge(df_employees, df_departments, on='EmployeeID')

# Display the result
print(merged_df)
````

````python
# Left join (all from left, matching from right)
pd.merge(df_employees, df_departments, on='EmployeeID', how='left')
````

````python
# Right join (all from right, matching from left)
pd.merge(df_employees, df_departments, on='EmployeeID', how='right')
````

````python
# Outer join (all rows from both)
pd.merge(df_employees, df_departments, on='EmployeeID', how='outer')
````

````python
# Label sessions that burned more than 400 calories
# If 'Calories' > 400, label as 'High', otherwise 'Normal'
data6['CalorieLevel'] = data6['Calories'].apply(
    lambda x: 'High' if x > 400 else 'Normal'
)

# Display the first few rows of the updated DataFrame
print(data6.head())
````

### ðŸ” Understanding `lambda x: 'High' if x > 400 else 'Normal'` in pandas

This expression is a **lambda function** used to classify values based on a condition. It is commonly used with `apply()` in pandas to create new columns.

#### âœ… What is a lambda function?

A `lambda` function is a quick, anonymous way to define a small function without giving it a name.

The general syntax is:

```python
lambda parameter: expression
```

It behaves like:

```python
def function(parameter):
    return expression
```

#### ðŸ”Ž Full Expression Breakdown:

```python
lambda x: 'High' if x > 400 else 'Normal'
```

| Part         | Explanation |
|--------------|-------------|
| `lambda x`   | Defines a function that takes a single input `x` (each value from the column) |
| `x > 400`    | Checks if the value is greater than 400 |
| `'High'`     | Returned if the condition is **True** |
| `'Normal'`   | Returned if the condition is **False** |

#### âœ… Example Use in pandas:

```python
data6['CalorieLevel'] = data6['Calories'].apply(lambda x: 'High' if x > 400 else 'Normal')
```

- `data6['Calories']`: A column of numeric values representing calories burned.
- `.apply(...)`: Applies the lambda function to **each individual value** in that column.
- Result: Creates a new column `'CalorieLevel'` with labels `'High'` or `'Normal'`.

#### ðŸ§ª Example:

| Calories | CalorieLevel |
|----------|---------------|
| 340.0    | Normal        |
| 409.1    | High          |
| 1500.0   | High          |

#### âœ… Why use it?

- Simple and readable for short, one-line conditions.
- Very useful for **data transformation**, **labeling**, and **conditional logic** in pandas.

````python
# Calculate pulse intensity (average pulse divided by duration)
# This gives an idea of heart rate effort per minute
data6['PulsePerMinute'] = data6.apply(
    lambda row: row['Pulse'] / row['Duration'], axis=1
)

# Display the first few rows of the updated DataFrame
print(data6.head())
````

````python
# Format the Calories column as a string with 'kcal' appended
# This is useful for display purposes (e.g., UI, reports, exports)
data6['CaloriesLabel'] = data6['Calories'].map(
    lambda x: f"{x:.3f} kcal"  # Format the number with 1 decimal place and add ' kcal'
    # For example, if x = 409.1 â†’ "409.100 kcal"
)

# Show the first few rows of the DataFrame to verify the new column
print(data6.head())
````

## Data Cleaning:

**Empty Cells:**

Empty cells can potentially give you a wrong result when you analyze data.

- **Remove Rows:**

One way to deal with empty cells is to remove rows that contain empty cells.

This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.

````python
# Return a new Data Frame with no empty cells:

data7 = pd.read_csv('data/data.csv')

data8 = data7.dropna()

print(data8)
````

Note: By default, the dropna() method returns a new DataFrame, and will not change the original.

If you want to change the original DataFrame, use the inplace = True argument:

````python
# Remove all rows with NULL values:

data9 = pd.read_csv('data/data.csv')

data9.dropna(inplace = True)

print(data9)
````

Note: Now, the dropna(inplace = True) will NOT return a new DataFrame, but it will remove all rows containing NULL values from the original DataFrame.

- **Replace Empty Values:**

Another way of dealing with empty cells is to insert a new value instead.

This way you do not have to delete entire rows just because of some empty cells.

The fillna() method allows us to replace empty cells with a value:

````python
# Replace NULL values with the number 130:

data10 = pd.read_csv('data/data.csv')

data10.fillna(130, inplace=True)

print(data10)
````

- **Replace Only For Specified Columns:**

The example above replaces all empty cells in the whole Data Frame.

To only replace empty values for one column, specify the column name for the DataFrame:

````python
# Replace NULL values in the "Calories" columns with the number 130:

data11 = pd.read_csv('data/data.csv')

data11.fillna({"Calories": 130}, inplace=True)

print(data11)
````

- **Replace Using Mean, Median, or Mode:**

A common way to replace empty cells, is to calculate the mean, median or mode value of the column.

Pandas uses the mean(), median() and mode() methods to calculate the respective values for a specified column:

````python
# Calculate the MEAN, and replace any empty values with it:

data12 = pd.read_csv('data/data.csv')

x = data12["Calories"].mean()

data12.fillna({"Calories": x}, inplace=True)
````

Mean = the average value (the sum of all values divided by number of values).

````python
# Calculate the MEDIAN, and replace any empty values with it:

data13 = pd.read_csv('data/data.csv')

x = data13["Calories"].median()

data13.fillna({"Calories": x}, inplace=True)
````

Median = the value in the middle, after you have sorted all values ascending.

````python
# Calculate the MODE, and replace any empty values with it:

data13 = pd.read_csv('data/data.csv')

# Calculate the mode (most frequent value) of the "Calories" column
# .mode() returns a Series; [0] selects the first mode in case there are multiple
x = data13["Calories"].mode()[0]

data13.fillna({"Calories": x}, inplace=True)
````

Mode = the value that appears most frequently.
