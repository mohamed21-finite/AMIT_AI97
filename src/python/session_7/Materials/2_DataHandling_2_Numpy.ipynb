# NumPy

**What is NumPy?**

NumPy (Numerical Python) is the core library for numerical computing in Python. It provides:

Fast multi-dimensional arrays

Mathematical operations on arrays

Linear algebra, statistics, and random number tools

**Why NumPy?**

Much faster than Python lists

Memory efficient

Foundation for Pandas, SciPy, Scikit-learn, TensorFlow

```python
# --- LINSPACE vs ARANGE ---

# Use linspace when you know HOW MANY points you want.
# "Give me exactly 5 numbers between 0 and 1 (includes 1.0)"
# Result: [0.  , 0.25, 0.5 , 0.75, 1.  ]
# np.linspace(0, 1, 5)

# Use arange when you know the STEP SIZE you want.
# "Give me numbers from 0 up to 1, jumping by 0.25 (excludes 1.0)"
# Result: [0.  , 0.25, 0.5 , 0.75]
# np.arange(0, 1, 0.25)

# KEY RULE:
# Linspace = Count-based (includes the end)
# Arange   = Step-based  (excludes the end)
```

## Broadcasting

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Broadcasting: The smaller array [10, 20, 30] is added to EVERY row of the matrix
# Row 1: [1+10, 2+20, 3+30] -> [11, 22, 33]
# Row 2: [4+10, 5+20, 6+30] -> [14, 25, 36]
# Result: [[11, 22, 33], [14, 25, 36]]
arr + np.array([10, 20, 30])
```

## Performance Comparison (Loop vs NumPy)

```python
import time
```

```python
size = 1_000_000
python_list = list(range(size))
np_array = np.arange(size)
```

```python
# Record the current time in seconds (the "start" timestamp)
start = time.time()
# A list comprehension: doubles every number in a standard Python list
[x * 2 for x in python_list]
# Subtract the start time from the current time to get the "elapsed time"
# This tells you exactly how many seconds the operation took
time.time() - start
```

```python
start = time.time()
np_array * 2
time.time() - start
```
