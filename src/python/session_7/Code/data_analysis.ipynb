```python
# DataFrame


data1 = {"calories": [420, 380, 390, 400, 500],
          "duration": [10, 30, 50, 60, 70]}

df = pd.DataFrame(data1)

print(df)
```

```python
print(df.loc[1])
```

```python
print(df.loc[[0,2]])
```

```python
print(df.loc[[1,3]])

print(df.loc[1:3])
```

```python
data2 = {
    "Name": ["Ahmed", "Mohamed", "Hossam", "Heba", "Mostafa"],
    "Age": [25, 30, 35, 40, 28],
    "City": ["Cairo", "Alex", "Giza", "Luxor", "Aswan"]
}

df2 = pd.DataFrame(data2, index=["a", "b", "c", "d", "e"])

print(df2)
```

```python
print(df2.loc["a"])

print(df2.loc['a':'d'])

print(df2.loc['c','Age']) # select row and column

print(df2.loc[['b', 'd'], ["Name", "City"]])
print(df2.loc['b': 'd', ["Name", "City"]]) # iloc works only for integer index
```

```python
print(df2.loc[df2["Age"] >= 30])
```
