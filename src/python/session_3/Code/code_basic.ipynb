```python
def word_length_sorter(word_list):
    word_length = {}
    for word in word_list:
        word_length[word] = len(word)

    for word, length in word_length.items():
        print(f"Word: {word}, Length: {length}", end=' ')
        print("long word") if length > 5 else print("short word")

word_length_sorter(['apple', 'banana', 'cherry', 'date', 'elderberry'])
```

```python
def lardest_number(num_list):
    largest = num_list[0]
    for num in num_list:
        if num > largest:
            largest = num
    print(f"The largest number is: {largest}")

lardest_number([12, 45, 7, 89, 23, 67, 34])
```

```python
def classify_numbers(n):
    divisible_by_3 = False
    divisible_by_5 = False

    for i in range(1, n+1):
        if i % 3 == 0:
            divisible_by_3 = True
        if i % 5 == 0:
            divisible_by_5 = True
    if divisible_by_3 and divisible_by_5:
        print("FizzBuzz")
    elif divisible_by_3:
        print("Fizz")
    elif divisible_by_5:
        print("Buzz")
    else:
        print(n)

classify_numbers(15)
```
