```python
def calculator(a,b,oper):
    if oper == '+':
        return a + b
    elif oper == '-':
        return a - b
    elif oper == '*':
        return a * b
    elif oper == '/':
        return a / b
    else:
        return "Invalid Operation"

try:
    a = float(input("enter first number: "))
    b = float(input("enter second number: "))
    op = input("enter operation: ")
    print(calculator(a , b, op))
except ZeroDivisionError:
    print("invalid number")
```

```python
grades_dict = {"Ahmed": [20, 20, 30], "Mohamed": [70, 35, 45], "Ali": [100, 20, 30]}

def grade_avg(grades):
    status = {}
    for i, j in grades.items():
        if sum(j) / len(j) >= 50:
            status[i] = "pass"
        else:
            status[i] = "fail"
    return status

print(grade_avg(grades_dict))

        
```

```python
num_list = [10, 101, 100, 201, 205, 107, 110, 20, 40, 55, 5]

def count_special_numbers(numbers):
    count = 0
    for i in numbers:
        if i > 100 and i % 2 != 0:
            count += 1
    
    return count

print(count_special_numbers(num_list))
```
