```python
class Animal:
    def speak(self):
        return "Animal speaks"

class Dog(Animal):
    def bark(self):
        return "Dog barks"

```

```python
class Flyer:
    def fly(self):
        return "Can fly"

class Swimmer:
    def swim(self):
        return "Can swim"

class Duck(Flyer, Swimmer):
    pass

duck = Duck()
print(duck.fly())
print(duck.swim())
```

```python
class Grandparent:
    def say_hello(self):
        return "Hello from Grandparent"

class Parent(Grandparent):
    def say_hi(self):
        return "Hi from Parent"

class Child(Parent):
    def say_hey(self):
        return "Hey from Child"

c = Child()
b = Parent()
d = Grandparent()

print(c.say_hello())
```

```python
# Base class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

# Child class (inherits from Person)
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)  # Call the constructor of the parent class
        self.student_id = student_id

    def study(self):
        return f"{self.name} is studying. Student ID: {self.student_id}"

# Creating objects
person1 = Person("Ali", 40)
student1 = Student("Mona", 20, "S123")

# Using methods
print(student1.study())
#print(student1.greet())     # Inherited from Person
#print(student1.study())     # Defined in Student
```

```python
class Vehicle():
    def __init__(self, name, model):
        self.name = name
        self.model = model
    
    def vehicle_info(self):
        return f"name: {self.name}, model: {self.model}"

class Car(Vehicle):
    def __init__(self,name, model, doorsnumber):
        super().__init__(name, model)
        self.doorsnumber = doorsnumber
    
    def car_info(self):
        return f"model: {self.model}, No.doors: {self.doorsnumber}"

vehicle1 = Vehicle("Bus", 2002)
vehicle2 = Vehicle("Truck", 2000)

car1 = Car("TOYOTA", 2023, 4)
car2 = Car("BMW", 2022, 4)

print(vehicle1.vehicle_info())
print(vehicle2.vehicle_info())
print()
print(car1.car_info())
print(car2.car_info())
```
