```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../Materials/city_sustainability_geospatial.csv')
```

```python
df
```

```python
#population_vs_co2 = px.scatter(df, x="population_millions", y="co2_tons_per_capita")
population_vs_co2 = px.scatter(df, x="population_millions", y="co2_tons_per_capita", color='continent', size='avg_temp_c', hover_name='city',
hover_data=['renewable_energy_share', 'air_quality_index'])
population_vs_co2.show()
```

```python
continent_vs_population_data = df.groupby('continent')['population_millions'].sum().reset_index()
continent_vs_population_data
```

```python
continent_vs_population = px.bar(continent_vs_population_data, x='continent', y='population_millions')
continent_vs_population.show()
```

```python
pop_hist = px.histogram(df, x='population_millions', nbins=40, color='continent', marginal='box')
pop_hist.show()
```

```python
continent_co2 = px.box(df, x='continent', y='co2_tons_per_capita', color='continent', points='all')
continent_co2.show()
```

```python
top_population_cities = df.nlargest(20, 'population_millions') 
top_population_cities 
```

```python
top_population_cities_costal = top_population_cities[top_population_cities['is_coastal'] == 'Yes']
top_population_cities_costal
```

```python
coastal_population = px.pie(top_population_cities, values='population_millions', names='is_coastal', hole=0.7)
coastal_population.show()
```

```python
fig = px.scatter_geo(df, lat='latitude', lon='longitude', color='continent', size='population_millions', hover_name='city'
                     , projection='natural earth', title='Global City Population Distribution', hover_data={'population_millions': True,
                                                                                                            'latitude': False, 'longitude': False,
                                                                                                            'continent': True, 'country': True})
fig.show()
```

```python
avg_co2_per_country = df.groupby('country')['co2_tons_per_capita'].mean().reset_index()
avg_co2_per_iso = df.groupby('iso_alpha3')['co2_tons_per_capita'].mean().reset_index()
avg_co2_per_country['iso_alpha3'] = avg_co2_per_iso['iso_alpha3']
avg_co2_per_country
```

```python
country_stats = df.groupby(['country', 'iso_alpha3'], as_index=False).agg(avg_co2=('co2_tons_per_capita', 'mean'),
                                                                         avg_temp=('avg_temp_c', 'mean'))
country_stats
```

```python
px.choropleth(
    country_stats,
    locations='iso_alpha3',
    color='avg_co2',
    hover_name='country',
    hover_data=['avg_temp'],
    color_continuous_scale='haline',
    projection='natural earth'
)
```
