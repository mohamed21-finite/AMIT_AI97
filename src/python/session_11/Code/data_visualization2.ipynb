```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../Materials/medical_data.csv')
```

```python
df
```

```python
plt.figure()
sns.histplot(df['age'], kde=True)
plt.show()
```

```python
plt.figure()
sns.kdeplot(df['systolic_bp'])
plt.show()
```

```python
plt.figure()
sns.boxplot(df['glucose'])
plt.show()
```

```python
plt.figure()
sns.countplot(x = df['smoker'])
plt.show()
```

```python
plt.figure()
sns.countplot(x = df['gender'])
plt.show()
```

```python
plt.figure()
sns.scatterplot(x = df['cholesterol'], y = df['bmi'])
plt.show()
```

```python
plt.figure()
sns.boxplot(x = df['gender'], y = df['bmi'])
plt.show()
```

```python
plt.figure()
sns.barplot(x = df['gender'], y = df['bmi'])
plt.show()
```

```python
plt.figure()
sns.violinplot(x = df['gender'], y = df['bmi'])
plt.show()
```

```python
plt.figure()
sns.scatterplot(x=df['cholesterol'], y=df['bmi'], hue=df['gender'], style=df['gender'], markers={"Male":"s", "Female": "o"})
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Reds')
plt.show()
```
