```python
q1, q3 = dropped_df['age'].quantile([0.25, 0.75])
iqr = q3 - q1
lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr

print("Q1:", q1)
print("Q3:", q3)
print("IQR:", iqr)
print("Lower Bound:", lower)
print("Upper Bound:", upper)
dropped_df[(dropped_df['age'] < lower) | (dropped_df['age'] > upper)]
```
