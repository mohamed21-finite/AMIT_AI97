```python
import pymongo

from pymongo import MongoClient
from bson.objectid import ObjectId


# Define connection string

client = MongoClient("mongodb://localhost:27017/")
```

```python
# create database
db = client["CustomersDB"]
```

```python
# create collection
customers_col = db["Customers"]
```

```python
def add_customer(customer):
    try:
        # 1. Insert customer dictionary into customers collection
        result = customers_col.insert_one({
            "Name": customer["name"],
            "Age": customer["age"],
            "Gender": customer["gender"],
            "Region": customer["region"]
        })
        return True, str(result.inserted_id)
    except Exception as e:
        return False, str(e)
```

```python
new_customer = {
    "name": "Ahmed Fahmy",  # Customer's name
    "age": 35,              # Customer's age
    "gender": "Male",       # Customer's gender
    "region": "Cairo"       # Customer's region
}
success, message = add_customer(new_customer)
print("Success:", success)
print("Message:", message)
```

```python
def get_customers():
    try:
        customers = []
        for doc in customers_col.find():
            doc['_id'] = str(doc['_id'])
            customers.append(doc)
        return True, customers

    except Exception as e:
        return False, str(e)
```

```python
success, customers = get_customers()

if success:
    print("Total customers:", len(customers))
    for c in customers:
        print(c)
else:
    print("Error fetching customers:", customers)
```

```python
def update_customer(filter, new_values):
    try:

        new_values = { "$set": new_values}
        # 1. Insert customer dictionary into customers collection
        customers_col.update_one(filter, new_values)
        return True, 'updated successfully'
    except Exception as e:
        return False, str(e)
```

```python
def delete_customer(filter):
    try:
        # 1. delete customer dictionary into customers collection
        customers_col.delete_one(filter)
        return True, 'deleted successfully'
    except Exception as e:
        return False, str(e)
```

```python
def add_many_customers(customers):
    try:
        # 1. Insert customer dictionary into customers collection
        result = customers_col.insert_many(customers)
        return True, "customers added successfully"
    except Exception as e:
        return False, str(e)
```

```python
filter = {'Name': "Ahmed Fahmy"}
new_value = {'Age': 45}
success, message = update_customer(filter, new_value)
print("Success:", success)
print("Message:", message)
```

```python
success, message = delete_customer(filter)
print("Success:", success)
print("Message:", message)
```

```python
customers = [{"Name": "Ahmed Fahmy", "Age": 40, "Gender": "Male","region": "Cairo"},
             {"Name": "Saitama", "Age": 25, "Gender": "Male","region": "Tokyo"}]

success, message = add_many_customers(customers)
print("Success:", success)
print("Message:", message)
```
