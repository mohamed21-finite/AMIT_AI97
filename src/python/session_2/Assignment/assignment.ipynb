9. Check if a set is subset of another

```python
set1 = {1, 2}
set2 = {1, 2, 3, 4}

# TODO: check subset manually or using issubset()
if set1.issubset(set2): # method 1
    is_subset = True

for i in set1:          # method 2
    if i not in set2:
        is_subset = False
        break
else:
    is_subset = True

print(is_subset)

# Hint: Use .issubset() or compare elements manually.
```

17. Replace the last value of each tuple in a list

- Sample list:
- [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
- Expected output:
- [(10, 20, 100), (40, 50, 100), (70, 80, 100)]

```python
data = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]

# TODO: replace last value with 100
new_list = []
new_list = list(data)
for i in range(len(new_list)):
    temp = list(new_list[i])
    temp[-1] = 100
    new_list[i] = tuple(temp)

print(new_list)
# Hint: Convert each tuple to list, edit, convert back.
```
