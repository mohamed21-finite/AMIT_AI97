```python
def sum1(x,y):
    '''
    param 1: take value from user as integer
    type(param-1): int
    param 2: take value from user as integer
    type(param-2): int

    return: make sum of param 1 and param 2
    type(return): int
    '''
    return x + y

def sub1(x,y):
    '''
    param 1: take value from user as integer
    type(param-1): int
    param 2: take value from user as integer
    type(param-2): int

    return: make sub of param 1 and param 2
    type(return): int
    '''
    return x - y

def mult1(x,y):
    '''
    param 1: take value from user as integer
    type(param-1): int
    param 2: take value from user as integer
    type(param-2): int

    return: make multipliction of param 1 and param 2
    type(return): int
    '''
    return x * y

def div1(x,y):
    '''
    param 1: take value from user as integer
    type(param-1): int
    param 2: take value from user as integer
    type(param-2): int

    return: make division of param 1 and param 2
    type(return): int
    '''
    return x // y

```

```python
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

sum_num = sum1(x, y)
sub_num = sub1(x, y)
mult = mult1(x, y)
div = div1(x, y)

print("sum: ", sum_num)
print("subtraction: ", sub_num)
print("multiplication: ", mult)
print("division: ", div)
```

```python
def even_odd(x):
    '''
    param 1: take value from user as list of integers
    type(param-1): list

    return: check number is even or odd
    type(return): two lists - even and odd
    '''
    odd = []
    even = []
    for i in x:
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)
    return even, odd
```

```python
numbers = []
elements = int(input("Enter number of elements in the list: "))
for i in range(elements):
    num = int(input(f"Enter element {i + 1}: "))
    numbers.append(num)
even_list, odd_list = even_odd(numbers)
print("Even numbers: ", even_list)
print("Odd numbers: ", odd_list)
```
