```python
degree = int(input("Enter your degree: "))

if degree < 0 or degree > 100:
    print("Please enter a valid degree between 0 and 100.")
else:
    if degree >= 90 and degree <= 100:
        print("Grade A")
    elif degree >= 80 and degree < 90:
        print("Grade B")
    elif degree >= 70 and degree < 80:
        print("Grade C")
    elif degree >= 60 and degree < 70:
        print("Grade D")
    else:
        print("Grade F")
```
