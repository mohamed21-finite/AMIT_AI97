# Task1

```python
s = input("Enter your email address: ")
extension_dict = {'com': 'Commercial Domain', 'edu': 'Educational Domain', 'org': 'Organizational Domain', 'net': 'Network Domain'}

if s.count('@') == 1 and s.count('.') >= 1 and s.startswith('@') is False and s.endswith('.') is False:
    username = s[:s.index('@')]
    domain = s[s.index('@'):].split('.')[0][1:]
    extension = s[s.index('@'):].split('.')[1][0:]

    print(f"Username: {username}")
    print(f"Domain: {domain}")
    print(f"Extension: {extension_dict.get(extension, 'Other Domain')}")
else:
    print("Invalid email format")
```

# Task2

```python
s = "###!!@moclew EPGTQ!!!6789"
first_part = s.split(' ')[0][s.index('@')+1:]
second_part = s.split(' ')[1]
second_part = second_part.replace('!!!6789', '')

first_part = list(first_part)
first_part.reverse()
first_part = ''.join(first_part)

second_part = second_part.replace('E', '')
decoded_message = first_part + ' ' + second_part
print("Decoded message:", decoded_message)
```

# TASK3

```python
encoded_message = '&&&**$gnirts PLIO!!@1234'


first_part = encoded_message.split(' ')[0][encoded_message.index('$')+1:]
second_part = encoded_message.split(' ')[1]
second_part = second_part.replace('!!@1234', '')
first_part = list(first_part)
first_part.reverse()
first_part = ''.join(first_part)

second_part = second_part.replace('I', 'E')
second_part = second_part.replace('O', 'U')
decoded_message = first_part + ' ' + second_part
print("Decoded message:", decoded_message)
```
